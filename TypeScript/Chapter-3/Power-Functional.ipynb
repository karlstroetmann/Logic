# A Functional Implementation of  `power`

The function `range` returns the list `[1, ..., n]`. 

```typescript
function range(n: number): number[] {
    return Array.from({ length: n }, (_, i) => i + 1);
}
```

```typescript
function power(m: bigint, n: number): bigint {
    return range(n).reduce((acc, _current) => acc * m, 1n); 
}
```

```typescript
console.log(power(3n, 4)); // Output: 81n
```
