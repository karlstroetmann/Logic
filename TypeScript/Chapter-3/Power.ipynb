# Computation of Powers via Iterative Squaring

The *recursive implementation* of the function `power` given below is based on the following two equations:
1. $m^0 = 1$
2. $m^n = \left\{\begin{array}{ll}
         m^{n/2} \cdot m^{n/2}                  & \mbox{if $n$ is even};    \\
         m^{(n-1)/2} \cdot m^{(n-1)/2} \cdot m  & \mbox{if $n$ is odd}.
         \end{array}
         \right.
         $
         
Observe that we square the result `p` of the recursive invocation of `power`. 
Hence this method is also known as *iterative squaring*.

As the powers of numbers get big fast, we use the type `bigint` for the base.
It would not make much sense to use this type for the exponent, since if we really have an exponent $n$ such that $n > 2^{53} - 1$, then $m^n$ would not fit into the computer memory unless $m$ is an element for the set $\{-1,0, 1\}$. 

```typescript
function power(m: bigint, n: number): bigint {
    if (n == 0) {
        return 1n;
    }
    const p = power(m, Math.floor(n / 2));
    if (n % 2 == 0) {
        return p * p;
    } else {
        return p * p * m;
    }
}
```

```typescript
for (let k = 0; k <= 10; ++k) {
    let kStr = k.toString().padStart(2);
    let pStr = power(3n, k).toString().padStart(6);
    console.log(`3 ** ${kStr} = ${pStr}`);
}
```

```typescript
console.time("power(3n, 500_000)");
const p = power(3n, 500_000);
console.timeEnd("power(3n, 500_000)");
```
