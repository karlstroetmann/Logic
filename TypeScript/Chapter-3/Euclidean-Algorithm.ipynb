# The Euclidean Algorithm for Computing the Greatest Common Divisor 

The function `ggt`takes to natural numbers `a` and `b`as arguments and computes their *greatest common divisor*.

```typescript
function ggt(a: number, b: number): number {
    while (b != 0) {
        [a, b] = [b, a % b];
    }
    return a;
}
```

```typescript
ggt(120, 20)
```
