# The Euclidean Algorithm for Computing the Greatest Common Divisor 

The function `gcd`takes to natural numbers `a` and `b`as arguments and computes their *greatest common divisor*.

```python
function gcd(a: int, b: int) -> int:
    while b != 0:
        a, b = b, a % b
    return a
```

```python
gcd(125, 20)
```
