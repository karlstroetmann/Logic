## Basic Concepts: Type Annotations and Functions

Before introducing the individual data types and language constructs of TypeScript, this section introduces two basic concepts that will recur throughout the chapter: type annotations and functions. In addition, the keywords `const` and `let` for variable declaration are introduced, which will be discussed in more detail later (see Section Primitive Data Types and Arithmetic).

**Type Annotations**

A central feature of TypeScript is the ability to explicitly assign a data type to variables, parameters, and return values. These so-called *type annotations* are noted with a colon (`:`) after the respective identifier:

```typescript
let name:   string  = "Alice"; // Variable of type string
let age:    number  = 25;      // Variable of type number
let active: boolean = true;    // Variable of type boolean
```

TypeScript also supports so-called *type inference*: If a value is assigned directly during declaration, the compiler automatically derives the type in most cases, making an explicit annotation optional.  For example, when we define the variable `email` below, TypeScript correctly infers that the value is a string.

```typescript
let email = "karl.stroetmann@dhbw.de"
```

**Variable Declaration with `const` and `let`**

To declare variables in TypeScript, the keywords `const` (for immutable variables) and `let` (for mutable variables) are used. Both are block-scoped, meaning they are only valid within the immediate block context (e.g., a function, a loop, or an if block) in which they were declared. The older keyword `var` is no longer recommended in modern TypeScript.

```typescript
const pi = 3.14159265;  // immutable
pi = 3;                 // Error: Assignment to constant variable.
```

```typescript
let counter = 0;  // mutable
counter = 1;      // allowed
```

**Functions**

A *function* is a named, reusable block of code that performs a defined task and optionally returns a value. In TypeScript, a function is declared using the `function` keyword. The types of the parameters as well as the return type are provided with annotations. The return type after the parameter list specifies what type the returned value must have. The `return` keyword terminates the execution of the function and returns the calculated value to the caller. If a function does not return a value, `void` is specified as the return type.

In the following code block, a function named *add* is defined, which takes two parameters *a* and *b* of type *number* and has *number* as a return value:

```typescript
function add(a: number, b: number): number {
    return a + b;
}
```

The function is called as follows:

```typescript
add(3, 5)
```

A function can also be defined using the `function` keyword in an expression.  The resulting function can then be given a name by assigning it to a variable, as shown below.

```typescript
let sub = function(a: number, b: number): number { 
    return a - b; 
};
```

```typescript
sub(3, 5)
```

**Arrow Functions**

In addition to the classic syntax, a more compact notation exists, the *arrow function syntax*. The `function` keyword is omitted; to the left of the arrow (`=>`) are the parameters, and to the right is the return expression:

```typescript
const add = (a: number, b: number): number => a + b;
```

```typescript
add(3, 5)
```

Arrow functions are often used in TypeScript for short, single-line expressions and are particularly used in combination with array methods like `filter`, `map`, and `reduce`, which are introduced in the Arrays section.

## Primitive Data Types and Arithmetic

In TypeScript, as in JavaScript, all numeric values are represented by default as floating-point numbers according to the [IEEE-754 standard](https://en.wikipedia.org/wiki/IEEE_754).  This applies even if the representation suggests an integer. The `typeof` operator returns the type of its operand as a string at runtime and is used for simple type inspection.

```typescript
typeof 42
```

```typescript
typeof 42.123
```

This design decision has direct implications for the precision of arithmetic operations. The precision for integers is limited by the constant `Number.MAX_SAFE_INTEGER`, which represents the value $2^{53} - 1$ (approx. 9 quadrillion). Beyond this limit, the exact representation of integers is no longer guaranteed, leading to rounding errors, as shown below.

```typescript
Number.MAX_SAFE_INTEGER
```

```typescript
Number.MAX_SAFE_INTEGER + 1 == Number.MAX_SAFE_INTEGER + 2
```

To bypass the limitations of the `number` type, the primitive data type `bigint` exists. It allows the representation of integers of arbitrary size, limited only by the available memory. Syntactically, `bigint` literals are denoted by the suffix `n`. Additionally, the wrapper constructor `BigInt()` is available, which converts a given value into the primitive type `bigint`, analogous to the relationship between `number` and the wrapper object `Number`.

An important aspect of type safety in TypeScript is the strict separation of these two numeric types. It is not possible to mix `bigint` and `number` in arithmetic operations without explicit conversion. Furthermore, the behavior during division differs: While `number` always yields floating-point results, the division of `bigint` values leads to a truncation of the decimal part (integer division).

```typescript
BigInt(1)
```

```typescript
const result: bigint = 7n ** 125n; // Calculation of extremely large powers
```

When values if type `bigint` are divided, the result is always of type `bigint` and hence is *truncated* sometimes.
Note that the reult is not **rounded** to the nearest integer, but rather it is *rounded towards zero*.

```typescript
19n/ 4n
```

```typescript
-19n / 4n
```

We can't mix the types `bigint` and `number` in arithemetic expressions. 

```typescript
2n * 3
```

In cases like these, we have to convert the `number` into a `bigint` using the function `BigInt`.

```typescript
2n * BigInt(3)
```

## Output and String Literals

For outputting values to the console, TypeScript provides the function `console.log`. This function is *polymorphic*, i.e. it accepts arguments of any type to output them in a textual representation. An elementary data type for textual data is the `string`. In TypeScript, string literals can be defined equivalently using single quotes (`'`) or double quotes (`"`). There is semantically no difference between these two notations, granting developers flexibility in adhering to coding style guidelines. Additionally, a third form of string literals exists, denoted by backticks (\`) and serving as the basis for so-called template strings, which will be discussed separately below.

In the interactive environment of a Jupyter Notebook, the result of the last expression of a cell is automatically output without an explicit call to `console.log` being necessary.

```typescript
// Explicit output via the console
console.log('1. Hello, World!');

// Alternative syntax with double quotes
console.log("2. Hello, World!")

// Implicit output of the last expression of a code cell
'Hello, World!'
```

**Template String**

In addition to classic string literals, TypeScript supports so-called *Template Strings*. These are enclosed by backticks "\`" and offer extended functionalities compared to regular strings. 
* A central feature is string interpolation, which allows expressions to be embedded directly into a string. 
  The syntax `${...}` is used here, where the contents of the curly braces are evaluated and the result is converted into 
  a *string*. This significantly increases the readability of the code, as complex concatenation operations are avoided. 
* Furthermore, template strings can span multiple lines without needing to insert special characters like `\n` into the code.

```typescript
const name = "Alice";
const age  = 25;

// Direct embedding of variables
console.log(`User ${name} is ${age} years old.`);

// Evaluation of expressions within the string
console.log(`Next year, she will be ${age + 1}.`);
```

## Arrays

Arrays are a fundamental data structure in TypeScript that represent a list of elements. They are conceptually equivalent to lists in Python, but differ in syntax and certain behaviors. An array is defined in TypeScript by square brackets "`[`" and "`]`". The empty array, i.e., an array without elements, is noted as follows: "`[ ]`".

An array with concrete elements is created by comma-separated values within the square brackets. Spaces after the commas are optional but recommended for better readability.

```typescript
[1, 2, 3]
```

**Accessing Array Elements: The Index Operator**

The *index operator* `[·]` is a postfix operator used to access the element at a specific position (index). TypeScript, like most modern programming languages, uses zero-based indexing, meaning the first element of an array has the index `0`. Note that `const` only prevents the reassignment of the variable, not the mutation of the array itself. The elements of the array can therefore still be modified:

```typescript
const arr = [1, 2, 3];
arr[0]; // Returns the first element
```

The index operator can also be used on the left side of an assignment to modify an existing element:

```typescript
arr[0] = 7;
arr
```

**Concatenating Arrays**

Arrays can be concatenated using the `concat` method. This method creates a new array containing all elements of the calling array followed by all elements of the array passed as an argument.

```typescript
[1, 2, 3].concat([4, 5, 6]); 
```

Alternatively, the *spread operator* "`...`" can be used to unpack and combine arrays. This syntax is often more compact when multiple arrays are concatenated and is frequently preferred in TypeScript.

```typescript
[...[1, 2, 3], ...[4, 5, 6]]
```

The `length` property returns the number of elements in an array.

```typescript
[4, 5, 4].length; 
```

## Functional Programming with Arrays

Arrays provide a number of methods that enable functional programming.  The methods are:
* `filter`,
* `map`, 
* `reduce`, 
* `some`, and `every`. 

**The `filter` Method**

The `filter` method creates a new array that contains only those elements of the original array for which a passed condition is met. The original array is not modified. The condition is passed as a function that returns either `true` (element is kept) or `false` (element is discarded) for each element. In TypeScript, the *arrow function syntax* (`=>`) is often used for such short functions.

In the following example, only numbers greater than 3 are filtered out of a list of numbers:

```typescript
const numbers = [1, 2, 3, 4, 5];
const result  = numbers.filter(n => n > 3);
result
```

The function `n => n > 3` is called internally for each element of the array. For `n = 1`, `2`, and `3`, it returns `false`, which is why these elements are discarded. For `n = 4` and `5`, it returns `true`, so they remain in the result array. The original array `numbers` remains unchanged.

| **Element (n)** | **Condition (n > 3)** | **Result** |
| :--- | :--- | :--- |
| 1 | false | Drop |
| 2 | false | Drop |
| 3 | false | Drop |
| 4 | true | Keep |
| 5 | true | Keep |

**The `map` Method**

The `map` method transforms each element of an array using a passed function and returns a new array containing the transformed values. The length of the resulting array is always identical to that of the original array.

In the following example, each element of a list of numbers is doubled:

```typescript
const numbers = [1, 2, 3, 4];
const result = numbers.map(n => n * 2); 
result
```

The function `n => n * 2` is called once for each element and returns its doubled value. The result is a new array where each position contains the transformed value of the corresponding position in the original. The original array `numbers` remains unchanged.

**The `reduce` Method**

The `reduce` method condenses an array into a single value by applying a function to each element along with a running intermediate result (the so-called *accumulator*).

The call `numbers.reduce((acc, n) => acc + n, 0)` consists of three parts:
1. The **initial value**: The `0` at the end is the initial value for the calculation.
2. The **callback function**: `(acc, n) => acc + n` defines the calculation rule.
3. The **parameters**:
    * `acc` (short for *Accumulator*): Stores the intermediate result of all previous additions.
    * `n`: Is the current element from the array that is currently being processed.

In the following example, the sum of all elements of an array is calculated:

```typescript
const numbers = [1, 2, 3, 4];
const sum = numbers.reduce((acc, n) => acc + n, 0);
sum
```

The accumulator `acc` starts with the value `0`. In each step, the current element `n` is added to the previous accumulator value, and the result is passed on as the new accumulator value:

| **Step** | **Accumulator (acc)** | **Element (n)** | **New acc** |
| :--- | :--- | :--- | :--- |
| Start | 0 | --- | 0 |
| 1 | 0 | 1 | 1 |
| 2 | 1 | 2 | 3 |
| 3 | 3 | 3 | 6 |
| 4 | 6 | 4 | 10 |

**The `some` Method**

The method `some` checks whether there is at least one element that satisfies a given condition. For example,
below we check whether `numbers` contains a number divisible by `13`.

```typescript
const numbers = [15, 30, 60, 69];
numbers.some(n => n % 13 == 0)
```

**The `every` Method**

The method `every` checks whether all elements of an array satisfy a given condition. For example,
below we check whether all numbers contained in `numbers` are even.

```typescript
const numbers = [2, 4, 6, 8, 10];
numbers.every(n => n % 2 == 0)
```

**Creating Arrays with `Array.from`**

Besides the literal notation (with square brackets `[ ]`), TypeScript offers the static method `Array.from` to dynamically create arrays. It is particularly useful when an array of a specific length needs to be populated with calculated values, for example, to generate a sequence of numbers.

The method accepts two arguments: first, an object with a `length` property that determines the desired length of the array, and an optional mapping function that is called for each index and determines the value at that position. Such a mapping function is so named because it systematically maps each original input value to a new, altered value by transforming it according to a defined rule and then storing it as a result in the new array. In the following example, an array with the numbers 1 to 10 is created:

```typescript
Array.from({length: 10}, (_, i) => i + 1);
```

The arguments are interpreted as follows:
1. `{length: 10}`: An object signaling `Array.from` to create an array of length 10. Since no concrete values are specified, each value is initially `undefined`.
2. `(_, i) => i + 1`: The mapping function determines the actual value at each position. It receives two arguments: the element itself (here `undefined`, since no value is present yet) and the current index `i`. Since the element is not needed, it is conventionally marked as *ignored* with an underscore (`_`). The index `i` runs from `0` to `9` and is transformed by `i + 1` into the desired values `1` to `10`.

**Array Destructuring**

A powerful feature of TypeScript is *Array Destructuring*. It allows unpacking values from an array directly into separate variables, rather than painstakingly accessing them via the index (like `arr[0]`, `arr[1]`). This leads to more compact and readable code, especially when functions return multiple values in the form of an array. TypeScript allows the declaration of a variable with `let` without an immediate value assignment. The value can be assigned at a later time.

```typescript
let a, b;
[a, b] = [10, 20];
console.log(a); 
console.log(b); 
```

Destructuring can also be combined with the *rest operator* (`...`) to collect the remaining elements of an array into a new array variable. This is particularly useful for separating the *head* and the *tail* of a list, a common pattern in functional programming.

```typescript
let head, tail;
[head, ...tail] = [10, 20, 30, 40, 50];
console.log(head); 
console.log(tail);
```

## Tuples

A *tuple* in TypeScript is a special array type that has a fixed number of elements with predefined data types. Unlike conventional arrays, which are mostly homogeneously typed (e.g., `string[]`), tuples allow the storage of a heterogeneous collection of values in a strict order. This concept is particularly suitable for grouping related data without having to define a complex object for it, for example for coordinate pairs `[x, y]` or return values of a function.

The type `boolean` denotes a truth value and can only take the two values `true` or `false`. These truth values are used in particular in conditions and control structures. A more detailed treatment of boolean values and operators is provided in the Boolean Values and Operators section.

In the following example, a tuple type `userProfile` is defined, which expects exactly three elements in the order `number`, `string`, and `boolean`:

```typescript
type UserProfile = [number, string, boolean]; // Type definition
let userProfile: UserProfile = [101, "Alice", true]; // Assignment: Order and types must match exactly
```

Access to the elements is analogous to arrays via the index operator:

```typescript
`Name: ${userProfile[1]}`
```

Many special constructs of the language, particularly arrays, are also based on objects at runtime but provide additional, list-like operations (e.g., `length` and array methods).

In TypeScript, object literals are defined with curly braces `{}`. The following example shows a `student` object with basic properties:

```typescript
console.log(student.firstName);               // Dot notation
const key = "matriculationNumber";
console.log(student[key]);                    // Bracket notation with variable
```

```typescript
student.isEnrolled = false;
```

## Boolean Values and Operators

The representation of truth values in TypeScript is done via the primitive data type `boolean`, which can only take the values `true` and `false`. These values are central for controlling the program flow and for logical decisions.

**Logical Operators and Lazy Evaluation**

TypeScript supports the classic logical operations conjunction (AND), disjunction (OR), and negation (NOT). The formal signatures of these operators look as follows:
* `&& (a: boolean, b: boolean) => boolean`
* `|| (a: boolean, b: boolean) => boolean`
* `! (a: boolean) => boolean`

An essential concept in the evaluation of `&&` and `||` is *lazy evaluation* (often referred to as *short-circuit evaluation* in this context). This means that the second operand is only evaluated if it is absolutely necessary for the final result:
* For `A && B`, `B` is no longer evaluated if `A` is already `false` (the result is definitely `false` anyway).
* For `A || B`, `B` is no longer evaluated if `A` is already `true` (the result is definitely `true` anyway).

```typescript
true && false; 
```

```typescript
true || false; 
```

```typescript
!true;       
```

**Comparison Operators and Strict Equality**

Comparisons are frequently used to generate truth values. A central concept in TypeScript is the distinction between *strict* and *loose* equality:
* **Strict Equality (`===` and `!==`):** Compares both the value and the data type. `5 === "5"` yields `false`, as a number is compared with a string. This is the recommended standard in TypeScript.
* **Loose Equality (`==` and `!=`):** In JavaScript these operators perform an automatic type conversion (*[Type Coercion](https://developer.mozilla.org/en-US/docs/Glossary/Type_coercion)*) before the comparison. Therefore, in JavaScript the expression `5 == "5"` yields `true`, since the string `"5"` is implicitly converted into the number 5.  This can lead to unexpected behavior and should be avoided.  However, this is not a problem in TypeScript, as the expression `5 == "5"` is not `well-typed` and hence is rejected by the TS compiler.

```typescript
5 == "5"
```

To my best knowledge, as long as no variable is typed as `any` (which essentially switches type checking off for this variable),
there is only one case where TypeScript allows implicit type conversions to go ahead.  The following example explains this case.

```typescript
null == undefined
```

```typescript
null === undefined
```

The classic size comparisons (`<`, `>`, `<=`, `>=`) are available in TypeScript.

An important difference compared to Python is the chaining of comparisons. A mathematical expression like $3 > 2 > 1$ is not accepted in TypeScript, because the result of `3 > 2` results in a `boolean` and we can't compare a `boolean` with a `number`.  Therefore, mathematical range checks must be explicitly linked with the *logical and operator*: `3 > 2 && 2 > 1`.

```typescript
3 > 2 > 1
```

```typescript
3 > 2 && 2 > 1
```

## Control Structures

Control structures serve to control the flow of a program. TypeScript offers classic constructs for this purpose, which are syntactically strongly based on C, Java, or C++.

**Branches**

The basic conditional execution is done using `if`, `else if`, and `else`. Unlike Python, the conditions must be placed in parentheses `()`. The associated instruction blocks are usually enclosed in curly braces `{}`. Although TypeScript syntax allows omitting the curly braces if the block consists of exactly one single instruction, explicit bracing is considered a best practice for readability and error prevention (e.g., when adding further instructions later).

```typescript
let x = 17;
if (x < 0) {
    console.log('The number is negative!');
} else if (x == 0) {
    console.log('The number is zero.');
} else {
    console.log("It's more than one.");
}
```

Alternatively, the `switch` statement offers a structured way to check multiple conditions against a value. At the end of each `case` block, the `break` keyword is used to force an exit from the `switch` structure. If `break` is missing, the program "falls through" into the next `case`. 

An interesting pattern in TypeScript is the `switch(true)` construct, which allows evaluating boolean expressions instead of static values in the `case` branches:

```typescript
switch (true) {
    case (x < 0):
        console.log('The number is negative!');
        break;
    case (x == 0):
        console.log('The number is zero.');
        break;
    default:
        console.log("It's more than one.");
}
```

**Loop Constructs**

TypeScript provides various loop constructs for iteration:
* The classic **`for` loop** uses a counter variable, a running condition, and a step size. It offers maximum control over the iteration process.

```typescript
const arr = [1, 2, 3];
for (let i = 0; i < arr.length; i++) {
    console.log(i);
}
```

* The **`for...of` loop** is ideally suited for iterating directly over the elements of an array or other iterable objects. It is more compact and less error-prone than the classic index-based `for` loop.

```typescript
const arr = [1, 2, 3];
for (const element of arr) {
    console.log(element);
}
```

* The **`while` loop** executes a block of code as long as a defined running condition is met.

```typescript
const arr = [1, 2, 3];
let i = 0;
while (i < arr.length) {
    console.log(arr[i]);
    i++;
}
```

Inside loops, the `continue` keyword can be used here to prematurely terminate the current loop iteration and continue with the next iteration.

```typescript
const arr = [1, 2, 3];
let i = -1;
while (i + 1 < arr.length) {
    ++i;
    if (i == 1) {
        continue;
    }
    console.log(`arr[${i}] = ${arr[i]}`);
}
```

**Application Example: [Sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)**

A classic algorithm for determining all prime numbers up to a given limit $n\in\mathbb{N}$ is the Sieve of Eratosthenes. The implementation in TypeScript demonstrates the interaction of arrays, the classic `for` loop as well as the `while` loop, and conditional statements:

1. Initialization of an array `primes`, where `primes[i] = i`.
2. Iteration over all numbers $i$ and $j$ to eliminate multiples (i.e., set them to 0).
3. Filtering the remaining numbers $\neq 0$.

An optimized version also uses `Math.ceil(Math.sqrt(n))` ($\lceil \sqrt{n} \rceil$) as the upper limit for the outer loop and starts the inner loop at $j=i$ to avoid redundant checks.

```typescript
function primes(n: number): number[] {
    // Initialize array with values 0 to n
    let primes = Array.from({length: n + 1}, (_, i) => i);
    let limit  = Math.ceil(Math.sqrt(n));
    for (let i = 2; i <= limit; i++) {
        // If i is already marked (0), skip
        if (primes[i] === 0) {
            continue; 
        }
        let j = i;
        // Cross out all multiples of i (starting at i*i)
        while (i * j <= n) {
            primes[i * j] = 0;
            ++j;
        }
    }
    // Filter all non-zero values (prime numbers)
    return primes.filter(p => p != 0 && p >= 2);
}    
```

Below, we need to use `JSON.stringify` to see the complete array.

```typescript
JSON.stringify(primes(10_000))
```

## Numerical Functions

TypeScript provides a comprehensive collection of standard mathematical functions and constants via the underlying JavaScript, encapsulated in the global `Math` object. This object acts as a namespace for static methods and properties. The elementary functions include:

**Rounding Functions**

Three specific methods are available for converting floating-point numbers into integers:
1. **`Math.floor(x)`:** The *floor function* $\lfloor x \rfloor$. It always rounds down to the next smaller integer.
$$ \text{floor}(x) = \max \{ n \in \mathbb{Z} \mid n \le x \} $$

```typescript
Math.floor(4.8)
```

$$ \text{ceil}(x) = \min \{ n \in \mathbb{Z} \mid n \ge x \} $$

```typescript
Math.ceil(4.2)
```

3. **`Math.round(x)`:** Commercial rounding to the nearest integer. Here, it rounds up starting from $.5$.

```typescript
Math.round(4.5)
```

* `Math.Pi = `$\pi$

   This is the ratio of a circle's perimeter to its diameter.
* `Math.E  = `$\displaystyle\sum\limits_{n=0}^{\infty} \frac{1}{n!}$ 

  This is [Euler's number](https://en.wikipedia.org/wiki/E_(mathematical_constant)).
