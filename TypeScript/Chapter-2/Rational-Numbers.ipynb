The function `gcd(n, m)` computes the *greatest common divisor* of `a` and `b`. 

````typescript
function gcd(a: number, b: number): number {
    if (b == 0) {
        return a;
    }
    return gcd(b, a % b);
}
````

````typescript
gcd(120, 100)
````

The function `generateRationals` is a *generator* that returns pairs of the form
`(n, q)` where `n` is a natural number and `q` is the $n^\mathrm{th}$ positive rational number.

* First, the rational number $\frac{0}{1}$ is generated.
* Second, all rational numbers $\frac{p}{q}$ are computed that satisfy $p > 0$, $p + q = 2$ and $\texttt{gcd}(p, q) = 1$,
* Third, all rational numbers $\frac{p}{q}$ are computed that satisfy $p > 0$, $p + q = 3$ and $\texttt{gcd}(p, q) = 1$,
* $\vdots$

````typescript
function* generateRationals(): Generator<[number, Fraction]> {
    let sum = 1;
    let cnt = 1;
    yield [cnt, new Fraction(0, 1)];
    sum += 1;
    cnt += 1;
    while (true) {
        for (let numerator = 1; numerator < sum; numerator++) {
            const denominator = sum - numerator;
            if (gcd(numerator, denominator) == 1) {
                yield [cnt, new Fraction(numerator, denominator)];
                cnt += 1;
            }
        }
        sum += 1;
    }
}
````

When we call `generateRational`, we get back an object `g` of type
```
Generator<[number, Fraction]>
```
This is a so called *iterator*.  This iterator has a method `next`.
Every time we call `g.next()` we will get an object of the form
$$ \{ \texttt{value}: [c, f], \texttt{done}: d\}, $$
where $c$ is a natural number that serves as a counter, while $f$ is an object that represents a *fraction*.  Finally, $d$ is a boolean value.  As long as there are any more rational numbers,  the value of $d$ will be `true`.  Since the set of rational numbers is infinite, $d$ will always be `true`.

````typescript
let g = generateRationals()
g
````

````typescript
let v1 = g.next()
v1
````

Here,  `{ s: 1n, n: 1n, d: 1n }` is an object representing the rational number
$$ s \cdot \frac{n}{d} $$
Therefore, $s$ represents the sign, $n$ represents the numerator, while $d$ is the denominator of the rational number.  If we want to write this rational number as a fraction, we can use the method `toFraction`.

````typescript
v1.value[1].toFraction()
````

The function `genRats(cnt)` prints the first `cnt` positive rational numbers.

````typescript
function gen(cnt: number): void {
    const g = generateRationals();
    for (let i = 0; i < cnt; i++) {
        const [index, frac] = g.next().value;
        let s1 = index.toString().padStart(3, ' ');
        let s2 = frac.toFraction().padStart(5, ' ')
        console.log(`${s1}: ${s2}`);
    }
}
````

````typescript
gen(100)
````
