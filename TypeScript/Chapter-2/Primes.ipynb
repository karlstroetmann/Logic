The function `countFrom(n)` creates the following *infinite stream* of natural numbers:
$$ [k \in \mathbb{N} \mid n \leq k] = [n,\; n+1,\; n+2,\; \cdots]$$

```typescript
function* countFrom(n: number): Generator<number> {
    while (true) {
        yield n++;
    }
}
```

The function `printFirst` prints the first `n` numbers produced by the generator `g`.

```typescript
function printFirst(g: Generator<number>, n: number): void {
    for (let i = 0; i < n; ++i) {
        console.log(g.next().value);
    }
}
```

```typescript
printFirst(countFrom(7), 3)
```

The function `filterStream` takes takes two arguments:
  * `g` is generator of numbers,
  * `p` is a predicate.
  
It returns a new generator that returns only those values `x` from `g`
such that `p(x)` is `true`.
$$ \texttt{filterStream}(G, p) = [ x \in G \mid p(x) ] $$

```typescript
function* filterStream(
   g: Generator<number>, 
   p: (n: number) => boolean
): Generator<number> {
    for (const x of g) {
       if (p(x)) {
           yield x;
       }
    }
}
```

Let us test `filterStream` by creating the a stream of numbers dividable by $3$.

```typescript
let divBy3 = filterStream(countFrom(1), n => n % 3 == 0);
printFirst(divBy3, 5)
```

The generator `filterPrime` takes a stream `g` of numbers, yields the first element `p`, filters all subsequent multiples of `p`, and calls `filterPrimes` recursively on the resulting stream.
$$ \texttt{filterPrimes}( [p, ... R]) = 
  \bigl[p, ...\texttt{filterPrimes}([ x \in R \mid x \,\texttt{%}\, p \not= 0])\bigr] $$

```typescript
function* filterPrime(g: Generator<number>): Generator<number> {
    const p = g.next().value; // get first number 
    yield p;                  // and return it, since it is prime
    // remove all multiples of prime p
    yield* filterPrime(filterStream(g, x => x % p != 0));
}
```

The function `primes` generates an infinite sequence of *prime numbers* using 
the [*Sieve of Eratosthenes*](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes).

```typescript
function* primes(): Generator<number> {
    yield* filterPrime(countFrom(2));
}
```

```typescript
printFirst(primes(), 100)
```
